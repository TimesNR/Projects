```python
def clasificar_fila(row):
    # Condiciones para que la vivienda sea considerada digna
    personas_por_cuarto = row['P1_1'] / row['P4_10'] <= 4
    tipo_de_vivienda = row['P4_3'] == 1 or 3 or 4 or 5
    material_paredes = row['P4_4'] >= 6
    material_techo = row['P4_5'] >= 6
    material_piso = row['P4_6'] >= 2
    cuartos_totales = row['P4_10A'] >= 6
    servicio_sanitario = row['P4_12'] == 1
    acceso_agua_llaves = row['P4_13'] == 1
    acceso_agua_entubada = row['P4_14'] == 1
    drenaje = row['P4_15'] == 1 or 2
    combustible_alimentos = row['P4_17'] == 1 or 3
    tiempo_construccion = row['P4_19_1'] <= 45
    metros_terreno = row['P4_20_1'] >= 75
    metros_construccion= (row['P4_21_1'] >= 28) & (row['P4_21_1'] <= 996)
    propietario_vivienda = row['P5_1'] == 1 or 4 or 5 or 6                                     
    
    if personas_por_cuarto and tipo_de_vivienda and material_paredes and material_techo and material_piso and cuartos_totales and servicio_sanitario and acceso_agua_llaves and acceso_agua_entubada and drenaje and combustible_alimentos and tiempo_construccion and metros_construccion and metros_terreno and propietario_vivienda:
        return 1
    else:
        return 0 # No digna

# Aplica la función de clasificación a cada fila del DataFrame
df_final['clasificacion'] = df_final.apply(clasificar_fila, axis=1)
```

```python
dt = pd.read_excel('a.xlsx')
num = 0
index = []
for i in range(len(dt['tabla'])):
    if dt['tabla'][i][2:3] != ',' :
        if int(dt['tabla'][i][1:3]) >5:
            index.append(i)
```

```python
num = 0
for vars in a:
    data = df_final[list(vars)] #, 
    data = data.dropna()
    data = data.to_numpy()
    pesos = [0]*np.shape(data)[0]
    for i in range(1,len(pesos)-1):
        pesos[i] = pesos[i]+4
    # Aqui es de np.shape = 99,2
    # por algun motivo solo le importa una dimension en este caso
    alpha_complex = gd.AlphaComplex(points=data, weights = pesos, precision = 'fast')
    simplex_tree = alpha_complex.create_simplex_tree()
    diag = simplex_tree.persistence()   
    #Plot the persistence diagram
    plt.figure()
    gd.plot_persistence_diagram(diag)
    plt.title("p "+vars[0] + ' ' + vars[1] +' '+vars[2])
    plt.savefig("p "+vars[0] + ' ' + vars[1] +' '+vars[2])
    plt.close(fig)
    aux = simplex_tree.persistent_betti_numbers(100,100)
    # print(aux)
    if aux != [1,0,0]: 
        vars_numerobety_diferente[num] = aux
        no_betty[num] = vars #--------------------------------------------------------<
    num +=1
    triangles = np.array([s[0] for s in simplex_tree.get_skeleton(2) if len(s[0]) == 3 and s[1]<=7.5])
    X = data.T[0]
    Y = data.T[1]
    Z = data.T[2]
    fig = plt.figure()
    ax = fig.add_subplot(projection='3d')
    plt.title('Alpha complex infinity data cloud')
    l = ax.plot_trisurf(X,Y,Z, triangles = triangles)
    plt.title("d "+vars[0] + ' ' + vars[1] +' '+vars[2])
    plt.savefig("d "+vars[0] + ' ' + vars[1] +' '+vars[2])
    plt.close(fig)
```
