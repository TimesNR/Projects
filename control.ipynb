```python
import numpy as np
import matplotlib.pyplot as plt
import random 
```

$ T(t) = \pm \ Ce^{-kt} + Tf$ 

```python
def T(t,C,k,Tf):
    return C * np.exp(-k*t) + Tf

def error(real, arr):
    error = [np.abs(real - v_calculado) for v_calculado in arr]
    return np.mean(error)

def optimizacion(N,C,Tf):
    aux = []
    k = np.random.uniform(1, 2, size=N)
    for i in range (N):
        ts = np.arange(0,10,0.1)
        Tt = [T(t,C,k,Tf)for t in ts]
        aux.append(error(Tf,Tt))
        indx_min = np.argmin(aux) 
        return k[indx_min]
```

```python
T0 = 20
Tf = 38 
ts = np.arange(0,10,0.01)
C = (Tf-T0)


k = optimizacion(100000,C,Tf)
Tt = [T(t,-C,k,Tf)for t in ts]
error(Tf,Tt)
```

```python
plt.figure(figsize=(12,8))
plt.title('Función de estabilización temperatura',fontsize=18)
plt.plot(ts,Tt, color= 'mediumvioletred',linewidth=2,label = '$T(t) = \pm \ Ce^{-kt} + Tf$')
plt.xlabel('Tiempo en seg',fontsize=12)
plt.ylabel('Temperatura en °C',fontsize=12)
plt.axhline(Tf,color = 'crimson', linestyle = '--', label = 'Temperatura final')
plt.axhline(T0,color = 'orchid', linestyle = '--', label = 'Temperatura inicial')
plt.grid(True)
plt.legend()
```
